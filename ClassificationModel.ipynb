```python
# Cleaned Dataset has the following information

# TRAINING DATA
# 'Text' : All the remaining words after the cleaning phase represented as a string
# 'Class' : The rating of the review for that specific entry (positive, negative, neutral)
# 'Text_As_Word_List' : All the remaining words after the cleaning phase represented as a list
# 'Text_As_Freq_Dict' : All the remaining words after the cleaning phase represented as a dictionary mapped to its frequency count

# TESTING DATA
# 'Text' : All the remaining words after the cleaning phase represented as a string
# 'Class' : NEED TO PREDICT
# 'Text_As_Word_List' : All the remaining words after the cleaning phase represented as a list
# 'Text_As_Freq_Dict' : All the remaining words after the cleaning phase represented as a dictionary mapped to its frequency count
```

```python
from ast import literal_eval
import pandas as pd
import numpy as np
```

```python
# Create dataframe back from the training datset file (need to use converter becuase list and dict is represented as stirng in csv)

df_train = pd.read_csv('cleaned_datasets/cleaned_training_data.csv', header='infer', low_memory=False, converters={'Text_As_Word_List' : literal_eval, 'Text_As_Freq_Dict' : literal_eval})
df_test = pd.read_csv('cleaned_datasets/cleaned_test_data.csv', header='infer', low_memory=False, converters={'Text_As_Word_List' : literal_eval, 'Text_As_Freq_Dict' : literal_eval})
```
